```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
plt.scatter(hd.chol,hd.target,marker='+',color='r')
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()

hd['chol']=sc.fit_transform(hd[['chol']])
hd['target']=sc.fit_transform(hd[['target']])
hd
```

```python
kval=range(1,10)
SSE=[]

for i in kval:
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(hd[['chol','target']])
    SSE.append(kmeans.inertia_)
plt.plot(kval,SSE) 
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
y_pred=kmeans.fit_predict(hd[['chol','target']])
y_pred
```

```python
hd['group']=y_pred
hd
```

```python
hd0=hd[hd['group']==0]
hd0
```

```python
hd1=hd[hd['group']==1]
hd1
```

```python
plt.scatter(hd1.chol,hd1.target,marker='*',color='green')
plt.xlabel('chol')
plt.ylabel('target')
```

```python
plt.scatter(hd0.chol,hd0.target,marker='*',color='green')
plt.scatter(hd1.chol,hd1.target,marker='*',color='red')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='*',color='black')
plt.xlabel('chol')
plt.ylabel('target')
```
